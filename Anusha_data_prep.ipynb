```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

data = {
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female'],
    'City': ['Mumbai', 'Delhi', 'Chennai', 'Mumbai', 'Delhi'],
    'Size': ['Small', 'Large', 'Medium', 'Large', 'Small'],
    'Age': [25, 45, 32, 60, 28],
    'Fare': [15, 300, 85, 450, 20]
}
df = pd.DataFrame(data)

# --- Task 1: Categorical Encoding ---

# 1. Gender -> Label Encoding
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])

# 2. City -> One-Hot Encoding (drop_first=True to avoid dummy variable trap)
df = pd.get_dummies(df, columns=['City'], drop_first=True)

# 3. Size -> Label Encoding with specific order (Manual Mapping)
# Standard LabelEncoder assigns alphabetically; we need to define the rank.
size_mapping = {'Small': 0, 'Medium': 1, 'Large': 2}
df['Size'] = df['Size'].map(size_mapping)

print("--- Encoded DataFrame ---")
print(df)
```

```python
from sklearn.preprocessing import RobustScaler

# --- Task 2: Feature Scaling ---

# Initialize RobustScaler
# We choose RobustScaler because it uses the median and IQR,
# preventing the 'Fare' outlier (450) from squashing the other data.
scaler = RobustScaler()

# Apply scaling to Age and Fare
df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']])

print("--- Final Scaled DataFrame ---")
print(df)

# Brief Justification:
# RobustScaler was chosen because the dataset contains a significant outlier (Fare=450).
# Unlike StandardScaler or MinMaxScaler, RobustScaler is not influenced by a few
# extreme values, ensuring the 'Age' and 'Fare' distributions remain meaningful.
```
